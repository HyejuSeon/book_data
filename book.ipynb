```python
from selenium import webdriver
from bs4 import BeautifulSoup

def org(year):
    url_end = ['A','C','D','E','I','J','K','N','S','a','b','c','e']
    org = []
    year = str(2019 - year)
    
    for i in range(len(url_end)):
        driver = webdriver.Chrome('C:/Users/tjswn/Downloads/chromedriver_win32/chromedriver.exe')
        driver.get('http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf?range=1&kind=3&orderClick=DAC&mallGb=KOR&linkClass=' + url_end[i])
        
        if url_end[i] == 'A':
            #연도선택
            temp_xp1 = '//*[@id="main_contents"]/div[3]/div/span[3]/a'
            temp_xp2 = '//*[@id="main_contents"]/div[3]/div/span[3]/ul/li[' + year + ']/a'
            #50개씩보기
            temp_xp3 = '//*[@id="main_contents"]/div[3]/div/span[1]/a'
            temp_xp4 = '//*[@id="main_contents"]/div[3]/div/span[1]/ul/li[2]/a'
            driver.find_element_by_xpath(temp_xp3).click()
            driver.find_element_by_xpath(temp_xp4).click()
        else:
            temp_xp1 = '//*[@id="main_contents"]/div[3]/div/span[2]/a'
            temp_xp2 = '//*[@id="main_contents"]/div[3]/div/span[2]/ul/li[' + year + ']/a'
        driver.find_element_by_xpath(temp_xp1).click()
        driver.find_element_by_xpath(temp_xp2).click()
        
        for page in range(1,3):
            if page == 1:
                xpath = '//*[@id="main_contents"]/div[4]/div[1]/ul/li[' + str(page) + ']/strong/a'
            else:
                xpath = '//*[@id="main_contents"]/div[4]/div[1]/ul/li[' + str(page) + ']/a'
            driver.find_element_by_xpath(xpath).click()
            
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            org.append(soup)
        driver.close()
    return org
```
